```python
hamletfile = open("hamlet.txt")
hamlettext=hamletfile.read()
hamletfile.close()
hamlettokens = hamlettext.split()
len(hamlettokens)
```

```python
uniquelctokens = set(hamlettokens)
print(uniquelctokens)
```

```python
tokensdict={}
for ut in uniquelctokens:
    tokensdict[ut] = hamlettokens.count(ut)
```

```python
from operator import itemgetter, attrgetter
L = sorted(tokensdict.items(), key=itemgetter(1),reverse=True)[:100]
print(L)
```

```python
topfreq = L[:20]

print(topfreq)

pos = np.arange(len(topfreq))

plt.bar(pos, [e[1] for e in topfreq])
plt.xticks(pos+0.4, [e[0 ] for e in topfreq])
```
